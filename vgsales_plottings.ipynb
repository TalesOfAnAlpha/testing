```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df=pd.read_csv("vgsales.csv")

```

SCATTER PLOT

1. It shows the relationship between datas of numerical value
2. The below graph shows the global sales of Nintendo published video games

```python
a= df.loc[df["Publisher"]=="Nintendo"]
sns.scatterplot(data=a ,x="NA_Sales",y="EU_Sales",hue="Global_Sales")
```

BARPLOT

1. barplot are used to compare between numerical and categorized data
2. The below graphs shows the games which are launched in their respective year

```python
b= df.head(3)
sns.barplot(data=b,x="Name",y="Year")
```

 DISTPLOT
 
 1. displot shows the variations in a data
 2. the below graph shows the variations in the EU_sales

```python
df2=df.head(150)
c=df2.EU_Sales
sns.displot(c)
```

LINEPLOT

In the below graph, the line plot is used to represent the loss and profit of the Nintendo published games

```python
sns.lineplot(data=a,x="Year",y="Global_Sales")
```
